# **Dimensionality Reduction & Feature Extraction via PCA**

```python
# Import necessary libraries
import itertools
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Olivetti faces dataset
faces = datasets.fetch_olivetti_faces()

# Print the dataset description
print(faces.DESCR)
```

```python
faces_images = faces['images']
faces_data  = faces.data
```

```python
faces_images.shape
```

```python
faces_data.shape
```

```python
# Create a figure with a specified size
fig = plt.figure(figsize=(16, 16))

# Plot the first 64 images in an 8x8 grid
for ii in range(64):
    plt.subplot(8, 8, ii + 1)
    plt.grid(False)
    plt.imshow(faces.images[ii], cmap=plt.cm.gray)
    plt.xticks([])
    plt.yticks([])

# Show the plot
plt.show()
```

```python
# Number of eigenfaces
n_eigenfaces = 16

# Create PCA object
pca = PCA(n_components=n_eigenfaces, whiten=True)
pca.fit(faces_data)

# Plot the eigenfaces
plt.figure(figsize=(16, 16))
plt.suptitle("Eigenfaces")
for ii in range(pca.components_.shape[0]):
    plt.subplot(4, 4, ii + 1)
    plt.imshow(pca.components_[ii].reshape(64, 64), cmap=plt.cm.gray)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

# Show the plot
plt.show()
```

```python
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(10, 4))
    plt.title('Explained Variance Ratio over Components')
    plt.plot(pca.explained_variance_ratio_)
    plt.xlabel('Component')
    plt.ylabel('Explained Variance Ratio')
    plt.show()
```

```python
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(10, 4))
    plt.title("Cumulative Explained Variance over Eigenfaces")
    plt.plot(pca.explained_variance_ratio_.cumsum())
    plt.xlabel('Component')
    plt.ylabel('Cumulative Explained Variance')
    plt.show()
```

```python
n_eigenfaces = 121

# Creating PCA object
pca = PCA(n_components=n_eigenfaces, whiten=True)
pca.fit(faces_data)

# Plot cumulative explained variance ratio
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(10, 4))
    plt.title("Cumulative Explained Variance over Eigenfaces")
    plt.plot(pca.explained_variance_ratio_.cumsum())
    plt.xlabel('Component')
    plt.ylabel('Cumulative Explained Variance')
    plt.show()
```
